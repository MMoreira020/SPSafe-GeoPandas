```python
#importando os objetos geométricos da biblioteca shapely
from shapely.geometry import Point, LineString, Polygon
```

```python
#criando uma geometria do tipo ponto
ponto = Point(5, 10)
ponto
```

```python
print(ponto)
type(ponto)
```

```python
#função coord
list(ponto.coords)
```

```python
#lendo x e y separadamente
x  = ponto.x
y = ponto.y
print(x, y)
```

```python
#calculando a distância entre os dois pontos
ponto_1 = Point(2,3)
ponto_2 = Point(5,7)

distancia = ponto_1.distance(ponto_2)

print("Distância entre os pontos", distancia)
```

Linha

```python
#criando uma linha a partir de objetos pontos
ponto_1 = Point(2,3)
ponto_2 = Point(5,7)
ponto_3 = Point(2,10)

linha = LineString([ponto_1, ponto_2, ponto_3])

linha
```

```python
print(linha)
type(linha)
```

Extraindo as coordenadas da LineString similar ao objeto Point

```python
#utilizando a função coords
list(linha.coords)
```

```python
#lendo x  e y separadamente
coordenadas_x = list(linha.xy[0])
coordenadas_y = list(linha.xy[1])

print("coordenadas x:", coordenadas_x)
print("coordenadas y:", coordenadas_y)
```

```python
#comprimento da linha
linha.length
```

```python
#obtendo o ponto central 
print(linha.centroid)
```

Polígono
Criação de polígonos => Pelo menos três coordenadas

```python
#criando um Polygon passando as coordenadas
poly = Polygon([(2,3), (5,7), (2,10)])
poly
```

```python
print(poly)
type(poly)
```

```python
#obtendo coordenadas do poligono
list(poly.exterior.coords)
```

```python
#Área
poly_area = poly.area
print("ÁREA = ", poly_area)
```

```python
# pegando o centroid do polígono
poly_centroid = poly.centroid
print("Centróide:", poly_centroid)
```

```python
#coordenadas do bounding box
poly.bounds
```

```python
#coordenadas da linha exterior
poly_exterior = poly.exterior

print("Exterior:", poly_exterior)
```

```python
#tamanho da linha exterior
poly_exterior_length = poly.exterior.length

print("Comprimento exterior:", poly_exterior_length)
```

# Operações básicas entre geometrias

```python
circulo_1 = Point(1,1).buffer(1)
circulo_2 = Point(2,1).buffer(1)

circulo_2
```

```python
import matplotlib.pyplot as plt
```

```python
x,y = circulo_1.exterior.xy
plt.plot(x,y)

x,y = circulo_2.exterior.xy
plt.plot(x,y)

plt.show()
```

# Intersecção

```python
circulo_inter = circulo_1.intersection(circulo_2)

circulo_inter
```

# União

```python
circulo_union = circulo_1.union(circulo_2)

circulo_union
```

# Diferença

```python
circulo_diff_1 = circulo_1.difference(circulo_2)
circulo_diff_1
```

```python
circulo_diff_2 = circulo_2.difference(circulo_1)
circulo_diff_2
```

# Diferença Simétrica

```python
circulo_diffsim = circulo_1.symmetric_difference(circulo_2)
circulo_diffsim
```
